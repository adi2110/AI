```
import pandas as pd

prev_app = pd.read_csv('previous_application.csv')
prev_app.head()
```

```
app_train = pd.read_csv('application_train.csv')
```

```
app_train.shape
```

```
import numpy as np
type(app_train)
```

```
app_train.head(50)
```

```
def find_missing(data):
    count = data.isnull().sum()
    total = data.shape[0]
    ratio = count/total
    
    return pd.DataFrame(data={'Count':count, 'Ratio':ratio}, index=data.columns.values)
find_missing(app_train)
```

```
categorical = app_train.select_dtypes(include='object').columns
categorical
```

```
numeric = app_train.select_dtypes(include=['int64', 'float']).columns
numeric
```

```
for col in categorical:
  app_train[col].fillna(app_train[col].mode()[0], inplace=True)
```

```
for col in numeric:
  app_train[col].fillna(app_train[col].mean(), inplace = True)
```

```
features = app_train.drop(['SK_ID_CURR', 'TARGET'], axis = 1)
target = app_train.TARGET
```

```
features.shape
```

```
target.shape
```

```
from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()
for i in categorical:
    features[i]=label.fit_transform(features[i])
```

```
features.head(30)
```

```
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(features, target, test_size=0.2)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
predictions = model.predict(X_test)

print(accuracy_score(y_test, predictions))
print(classification_report(predictions,y_test))
```

```
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier().fit(X_train,y_train)
predictions=model_rf.predict(X_test)

print(accuracy_score(predictions, y_test))
print(classification_report(predictions,y_test))
```

```
from xgboost import XGBClassifier

model_xgb = XGBClassifier(objective='binary:logistic')
model_xgb.fit(X_train,y_train)
predictions=model_xgb.predict(X_test)

print(accuracy_score(predictions, y_test))
print(classification_report(predictions,y_test))
```
