**DEPTH FIRST SEARCH**

```
class Node:
  def __init__(self, key):
    self.left = None
    self.right = None
    self.val = key 

def print_1(root):
  if root:
    print_1(root.left) 
    print(root.val), 
    print_1(root.right)

root = Node(1) 
root.left = Node(2) 
root.right = Node(3) 
root.left.left = Node(4) 
root.left.right = Node(5) 
root.right.left = Node(6)
root.right.right = Node(7)
root.left.left.left = Node(8)
root.left.left.right = Node(9)
root.left.right.left = Node(10)
root.left.right.right = Node(11)
root.right.left.left = Node(12)
root.right.left.right = Node(13)
root.right.right.left = Node(14)
root.right.right.right = Node(15)

print_1(root.right)
```

**Difference between Tensor and Multi-Dimensional Array**

1. A Tensor is a type of Function, whereas a Multi-dimensional array is a data structure suitable for representing a tensor

2. We need to specify transformation rules for the tensor but no specific transformations are needed for multidimensional arrays
